```python
import numpy as np
import pandas as pd
from tqdm import tqdm 
import matplotlib.pyplot as plt
```

```python
train_data = np.load("/data/train.npy")
test_data = np.load("/data/test.npy")
```

```python
train_distances = np.sum(np.sqrt(np.sum((train_data[:,:-1] - train_data[:,1:])**2,axis=-1)),axis=-1)
test_distances = np.sum(np.sqrt(np.sum((test_data[:,:-1] - test_data[:,1:])**2,axis=-1)),axis=-1)
```

```python
fig, axs = plt.subplots(2,1, sharex=True)
axs[0].hist(train_distances, bins=30)
axs[1].hist(test_distances, bins=30)
plt.show()
```
